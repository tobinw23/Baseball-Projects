```python
import pandas as pd
```

```python
df = pd.read_csv("/content/AnalyticsQuestionnairePitchData-1.csv")
df.head()
```

```python
display(df.head())
df.info()
display(df.isnull().sum())
```

```python
print(df.columns)
print(df.dtypes)
```

```python
print("Categorical columns:")
print(df.select_dtypes(include='object').columns)
print("\nNumerical columns:")
print(df.select_dtypes(include=['int64', 'float64']).columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='PitchType')
plt.title('Distribution of Pitch Types')
plt.xlabel('Pitch Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='PitchType', y='ReleaseSpeed')
plt.title('Release Speed Distribution by Pitch Type')
plt.xlabel('Pitch Type')
plt.ylabel('Release Speed')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='PitchCall')
plt.title('Distribution of Pitch Outcomes')
plt.xlabel('Pitch Call')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
pitch_call_by_type = pd.crosstab(df['PitchType'], df['PitchCall'])
display(pitch_call_by_type)
```
