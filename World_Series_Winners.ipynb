```python
from pybaseball import team_pitching
data = team_pitching(2013)
data.groupby('Team').get_group('BOS')
```

```python
data['Team'].unique()
```

```python
data1 = team_pitching(2013)
data2 = team_pitching(2014)
data3 = team_pitching(2015)
data4 = team_pitching(2016)
data5 = team_pitching(2017)
data6 = team_pitching(2018)
data7 = team_pitching(2019)
data8 = team_pitching(2020)
data9 = team_pitching(2021)
data10 = team_pitching(2022)
data11 = team_pitching(2023)
data2024 = team_pitching(2024)
```

```python
data_2013 = data1.groupby('Team').get_group('BOS')
data_2014 = data2.groupby('Team').get_group('SFG')
data_2015 = data3.groupby('Team').get_group('KCR')
data_2016 = data4.groupby('Team').get_group('CHC')
data_2017 = data5.groupby('Team').get_group('HOU')
data_2018 = data6.groupby('Team').get_group('BOS')
data_2019 = data7.groupby('Team').get_group('WSN')
data_2020 = data8.groupby('Team').get_group('LAD')
data_2021 = data9.groupby('Team').get_group('ATL')
data_2022 = data10.groupby('Team').get_group('HOU')
data_2023 = data11.groupby('Team').get_group('TEX')
```

```python
data2244=data2024.groupby('Team', as_index=False).apply(lambda x: x)
```

```python
combined_data = pd.concat([data_2013, data_2014, data_2015, data_2016, data_2017, data_2018, data_2019, data_2020, data_2021, data_2022, data_2023], ignore_index=True)
combined_data.dropna(how='all')
```

```python
print(combined_data.columns.to_list())
```

```python
import matplotlib.pyplot as plt

x = combined_data['Starting']
y = combined_data['Relieving']
labels = combined_data['Team'] + ' ' + combined_data['Season'].astype(str)

plt.figure(figsize=(10, 8))

plt.scatter(x, y)

for i, label in enumerate(labels):
    plt.annotate(label, (x[i], y[i]))

plt.axvline(combined_data['Starting'].mean(), color='black', linestyle='--')
plt.axhline(combined_data['Relieving'].mean(), color='black', linestyle='--')

plt.xlabel('Starting')
plt.ylabel('Relieving')
plt.title('Starting vs Relieving Pitching Stats (2013-2023)')

plt.show()
```
